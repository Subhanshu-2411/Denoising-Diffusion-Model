```python
def show_images(dataset, num_samples=20, cols=4):
  plt.figure(figsize=(15, 15))
  for i, image in enumerate(dataset):
    if i == num_samples:
      break
    plt.subplot(int(num_samples/cols) + 1, cols, i + 1)
    plt.imshow(image[0])
```

```python
data = torchvision.datasets.SVHN(root=".", download=True)
show_images(data)
```
